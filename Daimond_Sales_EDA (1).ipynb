##GRAPHICAL REPRESENTATIONS


#### HISTOGRAM

```python
plt.figure(figsize=(8,5))
plt.hist(df['PRICE'],bins=20,color='palegreen',edgecolor='k')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Diamond Prices')
plt.show()
```

#####  SCATTERPLOT

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='CARAT',y='PRICE',color='mediumturquoise',data=df)
plt.xlabel('carat Weight')
plt.ylabel('price')
plt.title('Carat Weight vs. Diamond Price')
plt.show()
```

#### BOX PLOT


```python
plt.figure(figsize=(8,5))
plt.subplot(1,2,1)
plt.boxplot(df['CARAT'])
plt.title("Carat Box Plot")

plt.subplot(1,2,2)
plt.boxplot(df['PRICE'])
plt.title('Price Box Plot')
plt.show()
```

```python
Q1 = df['CARAT'].quantile(0.25)
Q3 = df['CARAT'].quantile(0.75)
IQR = Q3 - Q1

# identify outliers
threshold = 1.5
outliers = data[(df['CARAT'] < Q1 - threshold * IQR) | (df['CARAT'] > Q3 + threshold * IQR)]
print(outliers.count())


```

```python
# handling outlier
z = np.abs(stats.zscore(df['CARAT']))
# replace outliers with median value
df.loc[z > threshold, 'CARAT'] = df['CARAT'].mean()
```

```python
Q1 = df['PRICE'].quantile(0.25)
Q3 = df['PRICE'].quantile(0.75)
IQR = Q3 - Q1

# identify outliers
threshold = 1.5
outliers = data[(df['PRICE'] < Q1 - threshold * IQR) | (df['PRICE'] > Q3 + threshold * IQR)]
print(outliers.count())
```

```python
# handling outlier
z = np.abs(stats.zscore(df['PRICE']))
# replace outliers with median value
df.loc[z > threshold, 'PRICE'] = df['PRICE'].mean()
```
